```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def histogram_plot(data, column, bins, kde, color, title, xlabel, ylabel):
    plt.figure(figsize=(8, 6))
    sns.histplot(data[column], bins=bins, kde=kde, color=color)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()


# Load the dataset
# Replace "your_dataset.csv" with the actual filename
data = pd.read_csv("Data_COVID19_Engl_sh_processed.csv")

# Explore the data
print(data.head())  # Display the first few rows of the dataset
# Display information about the dataset, including column names and data types
print(data.info())

# Descriptive statistics
print(data.describe())

# Data Visualization
# Histogram of Age Distribution
histogram_plot(data, 'age_n', 20, True, 'skyblue', 'Age Distribution', 'Age', 'Count')

# Bar plot of Marital Status Distribution
histogram_plot(data, 'mar_n', 4, False, 'lightgreen', 'Marital Status Distribution', 'Marital Status', 'Count')

# Bar Plot of sex_n
histogram_plot(data, 'sex_n', 3, False, 'yellow', 'Sex Distribution', 'Sex','Count')

# urban rural distribution
histogram_plot(data, 'urb_n', 2, False, 'lightcoral', 'Urban/Rural Distribution', 'Urban/Rural', 'Count')

# Education level distribution
histogram_plot(data, 'edu_n', 7, False, 'lightblue', 'Education Level Distribution', 'Education Level', 'Count')

#Income level distribution
histogram_plot(data, 'iin_n', 7, False, 'lightcoral', 'Income Level Distribution', 'Income Level', 'Count')

# sector 
histogram_plot(data, 'sec_n',7, False, 'lightcoral', 'Sector Distribution', 'Sector', 'Count')

#diagnosed with covid
histogram_plot(data, 'cov_n', 2, False, 'lightcoral', 'Diagnosed with COVID-19', 'Diagnosed with COVID-19', 'Count')
```
